```python
#1)A Agrupe os dados por tipo de bebidas;

import pandas as pd

#Importando Arquivo CSV para Leitura
df = pd.read_csv('world_alcohol.csv')

#declarando a variável
lista_bebidas = df.groupby("Beverage Types").size().reset_index(name='Quantidade')

#Imprimindo informações da lista de bebidas
print(lista_bebidas)


```

```python
#1) C Seção de Contagens: Contar a ocorrência de Regiões, de Países e a soma da coluna de valores por Bebida.


import pandas as pd


df = pd.read_csv('world_alcohol.csv')

# Contagem de ocorrência
contagem_regioes = df.groupby(['WHO region', 'Country']).size().reset_index(name='Contagem')
print("Ocorrência de Regiões e Países:")
print(contagem_regioes)

# Soma da coluna de valores por Bebida
somabebida = df.groupby('Beverage Types')['Display Value'].sum()
print("Soma por Bebida:")
print(somabebida)





```

```python
#1)D Realize análises estatísticas da coluna dos valores: Média, Moda, Mediana, Estatística
#Descritiva e Gráfico de comparação dos valores agrupados por tipo de bebida.

import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv('world_alcohol.csv')

# Análise estatística da coluna de valores
media = df['Display Value'].mean()
moda = df['Display Value'].mode().values[0]
mediana = df['Display Value'].median()
estatis = df['Display Value'].describe()

# Agrupar os valores por tipo de bebida
grupo_val = df.groupby('Beverage Types')['Display Value'].sum()

# Exibir análises estatísticas
print("Análises estatísticas da coluna de valores:")
print("Média:", media)
print("Moda:", moda)
print("Mediana:", mediana)
print("Estatística Descritiva:\n", estatis)

# Gráfico de comparação dos valores agrupados por tipo de bebida
grupo_val.plot(kind='bar', rot=0)
plt.xlabel('Tipo de Bebida')
plt.ylabel('Soma dos Valores')
plt.title('Comparação dos Valores por Tipo de Bebida')
plt.show()

```

```python
#Mostre resultados de acordo com alguns critérios:
#1 E)I. Mostrar a coluna de bebidas do ano de 1985.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('world_alcohol.csv')

# Filtrando o ano
df_year = df[df['Year'] == 1985]

# Exibindo a coluna de bebidas
coluna_bebidas_1985 = df_year['Beverage Types']

# Contar a ocorrência de cada tipo de bebida
contbebidas = coluna_bebidas_1985.value_counts()

# Construíndo gráfico para melhor visualização
contbebidas.plot(kind='bar', figsize=(8, 6))
plt.title('Ocorrência de Bebidas em 1985')
plt.xlabel('Bebidas')
plt.ylabel('Contagem')
plt.show()



```

```python
#Mostre resultados de acordo com alguns critérios:
#1 E)II. Mostrar a coluna de Região com valores acima de 4.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('world_alcohol.csv')

# Filtrar o DataFrame pelos valores acima de 4 na coluna 'WHO region'
df_filtrado = df[df['WHO region'] > '4']

# Exibir a coluna de 'WHO region' com valores acima de 4
colfiltrada = df_filtrado['WHO region']

# Plotar o gráfico de barras para visualização
plt.figure(figsize=(8, 6))
colfiltrada.value_counts().plot(kind='bar')
plt.title("Contagem de 'WHO region' com valores acima de 4")
plt.xlabel("WHO region")
plt.ylabel("Contagem")
plt.xticks(rotation=20)
plt.show()
```

```python
#2) I. Estatística Descritiva das Notas Integral Ampla dos cursos de Bacharelado.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('cursos-prouni.csv', encoding='UTF-8')

df_filtrado = df.loc[(df['grau'] == 'Bacharelado') & (df['nota_integral_ampla'].notna()), 'nota_integral_ampla']

if df_filtrado.empty:
    print("Não há dados de notas para cursos de bacharelado na modalidade Integral Ampla.")
else:
    notas = df_filtrado.astype(float)
    estatis = {
        'Média': np.mean(notas),
        'Mediana': np.median(notas),
        'Desvio Padrão': np.std(notas),
        'Mínimo': np.min(notas),
        'Máximo': np.max(notas),
    }

    # Criação do gráfico de barras
    plt.bar(estatis.keys(), estatis.values())

    # Personalização do gráfico
    plt.title('Estatísticas Descritivas das Notas Integral Ampla de Cursos de Bacharelado')
    plt.xlabel('Estatísticas')
    plt.ylabel('Valores')

    # Exibição do gráfico
    plt.show()


```
